```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/drive/My Drive/Screener/equitydata-screener-all-stocks-20250129.csv'
data = pd.read_csv(file_path)

# Filter relevant columns and drop missing values
columns_needed = ['Industry', 'RSI']
filtered_data = data[columns_needed].dropna()

# Group by Industry and calculate mean and median RSI
industry_rsi_stats = filtered_data.groupby('Industry')['RSI'].agg(['mean', 'median']).reset_index()

# Display the calculated statistics
print(industry_rsi_stats)

# Plotting Mean RSI by Industry
plt.figure(figsize=(12, 6))
plt.bar(industry_rsi_stats['Industry'], industry_rsi_stats['mean'])
plt.xticks(rotation=90)
plt.title('Average RSI by Industry')
plt.ylabel('Average RSI')
plt.xlabel('Industry')
plt.show()

# Plotting Median RSI by Industry
plt.figure(figsize=(12, 6))
plt.bar(industry_rsi_stats['Industry'], industry_rsi_stats['median'])
plt.xticks(rotation=90)
plt.title('Median RSI by Industry')
plt.ylabel('Median RSI')
plt.xlabel('Industry')
plt.show()
```

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the path to the folder containing sector-specific CSV files
folder_path = '/content/drive/My Drive/Screener/20241122/'
for file_name in os.listdir(root_folder): print(file_name)
subdirectory_path = os.path.join(root_folder, '20241122')

if os.path.exists(subdirectory_path):
    for file_name in os.listdir(subdirectory_path):
        print(file_name)
else:
    print(f"Subdirectory {subdirectory_path} does not exist.")
```

```python
# Initialize an empty DataFrame to store aggregated data
all_sector_data = pd.DataFrame()

# Iterate through all CSV files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):  # Process only CSV files
        file_path = os.path.join(folder_path, file_name)
        try:
            # Read each CSV file
            sector_data = pd.read_csv(file_path)

            # Check if 'Industry' and 'RSI' columns exist
            if 'Industry' in sector_data.columns and 'RSI' in sector_data.columns:
                all_sector_data = pd.concat([all_sector_data, sector_data[['Industry', 'RSI']]])
        except Exception as e:
            print(f"Error reading {file_name}: {e}")

# Drop missing values and calculate average RSI by industry
all_sector_data = all_sector_data.dropna()
industry_avg_rsi = all_sector_data.groupby('Industry')['RSI'].mean().reset_index()

# Prepare data for heatmap
industry_avg_rsi = industry_avg_rsi.sort_values('RSI', ascending=False)
heatmap_data = industry_avg_rsi.set_index('Industry')

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Industry Average RSI Heatmap')
plt.xlabel('RSI')
plt.ylabel('Industry')
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to the folder containing sector-specific CSV files
folder_path = '/content/drive/My Drive/Screener/20241122/'

# Initialize an empty list to store sector-wise PE averages
sector_pe_data = []

# Iterate through all CSV files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):  # Process only CSV files
        file_path = os.path.join(folder_path, file_name)
        try:
            # Read each CSV file
            sector_data = pd.read_csv(file_path)

            # Check if 'Price to Earning' column exists
            if 'Price to Earning' in sector_data.columns:
                # Calculate the average Price to Earning for the sector
                sector_avg_pe = sector_data['Price to Earning'].mean()
                # Extract the sector name (file name without extension)
                sector_name = os.path.splitext(file_name)[0]
                # Append to the list
                sector_pe_data.append({'Sector': sector_name, 'Average PE': sector_avg_pe})
            else:
                print(f"Skipped {file_name}: 'Price to Earning' column not found.")
        except Exception as e:
            print(f"Error reading {file_name}: {e}")

# Convert the list to a DataFrame
all_sector_data = pd.DataFrame(sector_pe_data)

# Check the contents of the DataFrame
print(all_sector_data.head())

# Ensure the DataFrame contains the required columns
if not all_sector_data.empty:
    # Sort data by Average PE
    all_sector_data = all_sector_data.sort_values(by='Average PE', ascending=False)

    # Plot the graph
    plt.figure(figsize=(12, 8))
    plt.bar(all_sector_data['Sector'], all_sector_data['Average PE'])
    plt.xticks(rotation=90)
    plt.title('Sector Average Price to Earning')
    plt.ylabel('Average Price to Earning')
    plt.xlabel('Sector')
    plt.tight_layout()
    plt.show()
else:
    print("No data to display. Check if the 'Price to Earning' column exists in the sector-specific files.")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to the folder containing sector-specific CSV files
folder_path = '/content/drive/My Drive/Screener/20241122/'

# Initialize an empty list to store sector-wise percentage of stocks
sector_percentage_data = []

# Iterate through all CSV files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):  # Process only CSV files
        file_path = os.path.join(folder_path, file_name)
        try:
            # Read each CSV file
            sector_data = pd.read_csv(file_path)

            # Check if 'Price to Earning' and 'Industry PE' columns exist
            if 'Price to Earning' in sector_data.columns and 'Industry PE' in sector_data.columns:
                # Calculate the percentage of stocks where Price to Earning < Industry PE
                total_stocks = len(sector_data)
                if total_stocks > 0:  # Avoid division by zero
                    stocks_below_industry_pe = (sector_data['Price to Earning'] < sector_data['Industry PE']).sum()
                    percentage_below_industry_pe = (stocks_below_industry_pe / total_stocks) * 100
                    # Extract the sector name (file name without extension)
                    sector_name = os.path.splitext(file_name)[0]
                    # Append the data
                    sector_percentage_data.append({'Sector': sector_name, 'Percentage Below Industry PE': percentage_below_industry_pe})
        except Exception as e:
            print(f"Error reading {file_name}: {e}")

# Convert the list to a DataFrame
sector_percentage_df = pd.DataFrame(sector_percentage_data)

# Sort the DataFrame in descending order
sector_percentage_df = sector_percentage_df.sort_values(by='Percentage Below Industry PE', ascending=False)

# Plot the graph
plt.figure(figsize=(15, 8))
plt.bar(sector_percentage_df['Sector'], sector_percentage_df['Percentage Below Industry PE'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Percentage of Stocks Below Industry PE by Sector', fontsize=16)
plt.ylabel('Percentage Below Industry PE (%)', fontsize=14)
plt.xlabel('Sector', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data
#directory_path = '/Users/sangram.dash2/Downloads/PD2025/incident_activity_export'
#csv_file = 'equitydata-screener-all-stocks-09APR.csv'
#df = pd.read_csv(os.path.join(directory_path, csv_file))

# Define the file path
file_path = '/content/drive/My Drive/Screener/equitydata-screener-all-stocks-20250409.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()


# Drop rows with missing RSI or Industry
rsi_df = df.dropna(subset=["RSI", "Industry"])

# Group by industry and calculate average RSI
industry_rsi = rsi_df.groupby("Industry")["RSI"].mean().sort_values(ascending=False)

# Convert to DataFrame for heatmap
industry_rsi_df = pd.DataFrame(industry_rsi).reset_index()
industry_rsi_df.columns = ["Industry", "Average RSI"]
industry_rsi_df = industry_rsi_df.set_index("Industry")

# Plot heatmap
plt.figure(figsize=(10, 20))
sns.heatmap(industry_rsi_df, annot=True, cmap="RdYlGn_r", linewidths=0.5,
            linecolor='gray', cbar_kws={'label': 'RSI'})
plt.title("Industry Average RSI Heatmap", fontsize=16, fontweight='bold')
plt.xlabel("RSI")
plt.ylabel("Industry")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from file
file_path = '/Users/sangram.dash2/Downloads/equitydata-screener-all-stocks-09APR.csv'
df = pd.read_csv(file_path)

# Drop rows with missing required values
industry_pe_df = df.dropna(subset=['Price to Earning', 'Industry PE', 'Industry']).copy()

# Classify each stock as above or below its Industry PE
industry_pe_df['Relative Position'] = industry_pe_df.apply(
    lambda row: 'Above Industry PE' if row['Price to Earning'] > row['Industry PE'] else 'Below Industry PE',
    axis=1
)

# Group by Industry and Relative Position, then reshape
industry_pe_counts = industry_pe_df.groupby(['Industry', 'Relative Position']).size().unstack(fill_value=0)

# Sort industries by total number of stocks
industry_pe_counts['Total'] = industry_pe_counts.sum(axis=1)
industry_pe_counts = industry_pe_counts.sort_values('Total', ascending=False).drop(columns='Total')

# Plotting
plt.figure(figsize=(16, 10))
industry_pe_counts.plot(
    kind='barh',
    stacked=True,
    color={'Above Industry PE': 'green', 'Below Industry PE': 'red'},
    edgecolor='black',
    figsize=(16, 10)
)

# Format the plot
plt.title('ðŸ“Š Stocks Above & Below Industry P/E by Industry', fontsize=16, fontweight='bold')
plt.xlabel('Number of Stocks', fontsize=12)
plt.ylabel('Industry', fontsize=12)
plt.legend(title='Position vs Industry PE', fontsize=10, title_fontsize=11)
plt.xticks(fontsize=10)
plt.yticks(fontsize=9)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.gca().invert_yaxis()  # Highest total at the top

# Display the plot
plt.show()
```
